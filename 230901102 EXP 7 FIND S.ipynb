```python
# FIND-S Algorithm Implementation

def find_s_algorithm(examples):
    
    # Step 1: Initialize hypothesis to most specific
    hypothesis = ['Ø'] * (len(examples[0]) - 1)

    # Step 2: Iterate through training examples
    for example in examples:
        
        # Consider only positive examples
        if example[1] == 'Warm':
            
            for i in range(len(hypothesis)):
                
                # If hypothesis is most specific, replace with attribute value
                if hypothesis[i] == 'Ø':
                    hypothesis[i] = example[i]
                
                # If values differ, generalize with '?'
                elif hypothesis[i] != example[i]:
                    hypothesis[i] = '?'
    
    return hypothesis


# Training Data
training_data = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Rainy', 'Cold', 'High', 'Strong', 'Warm', 'Change', 'No'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change', 'Yes']
]

# Run Algorithm
final_hypothesis = find_s_algorithm(training_data)

print("Final Hypothesis:", final_hypothesis)
```
